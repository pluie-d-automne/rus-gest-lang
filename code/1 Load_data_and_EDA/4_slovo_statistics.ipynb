```python
slovo_df['frames_cnt'] = slovo_df['end']-slovo_df['begin']+1
slovo_df.head()
```

```python
slovo_df.groupby(['height', 'width']).agg({'attachment_id' : ['count'], 'frames_cnt':['sum']}).sort_values(['height', 'width'])
```

```python
slovo_df['CLEANED'] = slovo_df.text.str.lower()\
    .str.replace("[.?!,:.\"–=#\^]", "", regex=True)
```

```python
slovo_df['LENGTH'] = slovo_df['CLEANED'].str.split(' ').str.len()
```
